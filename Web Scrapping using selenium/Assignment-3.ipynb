#### Ques 1 and Ques 2
1.Write a python program which searches all the product under a particular product from www.amazon.in. 
The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search 
for guitars.

2.In the above question, now scrape the following details of each product listed in first 3 pages of your 
search results and save it in a data frame and csv. In case if any product has less than 3 pages in search 
results then scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“

```python
url = 'https://www.amazon.in/' #opening the website
driver.get(url)

time.sleep(2)

search = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input') #given the xpath for the search button
search.send_keys(input())  #Used Input function so that each and every time user could give unique keyword
click = driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]') #given the xpath for the click button
click.click()
```

Here I have extracted the url's

```python
start = 0
end = 3 #Used 3 here, because we have to extract the data till page 3.

url = []
for i in range(start,end):
    try:
        ur = driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]/a')
        for i in ur:
            url.append(i.get_attribute('href'))       
    except NoSuchElementException:
        url.append("-")
    
    time.sleep(2)
    nxt_button = driver.find_elements_by_xpath("//li[@class='a-last']/a")   
    try:
        driver.get(nxt_button[0].get_attribute('href')) 
    except:
        driver.get(nxt_button[1].get_attribute('href'))
len(url)
```

```python
brand=[]      #Empty list
price=[]
product=[]
exchange=[]
delivery=[]
availability=[]

for i in url:    
    try:                           #Used try and except{Just in case if we don't have data in the particular class, then also code should run without giving error.}
        driver.get(i)
        time.sleep(2)              #Given the time to the browser, so that it could load the page completly.
        br=driver.find_element_by_xpath('//td[@class="a-span9"]/span')
        brand.append(br.text)                                  
    except NoSuchElementException:
        brand.append("-")
        
    try:
        pri=driver.find_element_by_xpath('//span[@class="a-price-whole"]')
        price.append(pri.text)
    except NoSuchElementException:
        price.append("-")
        
    try:
        pro=driver.find_element_by_xpath('//span[@class="a-size-large product-title-word-break"]')
        product.append(pro.text)
    except NoSuchElementException:
        product.append("-")
    
    try:
        exc=driver.find_element_by_xpath('//*[@id="RETURNS_POLICY"]/span/div[2]')
        exchange.append(exc.text)
    except NoSuchElementException:
        exchange.append("-")
    
    try:
        de=driver.find_element_by_xpath('//*[@id="ddmDeliveryMessage"]/b')
        delivery.append(de.text)
    except NoSuchElementException:
        delivery.append("-")
    
    try:
        ava=driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')
        availability.append(ava.text)
    except NoSuchElementException:
        availability.append("-")
        
print(len(brand))              #here I have checked the length of each data so that while creating a dataframe I don't face any error.
print(len(price)) 
print(len(product))
print(len(exchange))
print(len(delivery))
print(len(availability))
```

```python
# Creating the DataFrame - 
df = pd.DataFrame({'Brand Name':brand,'Name of the Product':product,'Price':price,'Return/Exchange':exchange,'Expected Delivery':delivery,'Availability':availability,'Product URL':url})
df
```

```python
# Saving the DataFrame as csv file for our future use.
df. to_csv("Product Details.csv")
```

## Ques. 4
Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on
www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be 
scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, 
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
url3 = 'http://www.flipkart.com' #opening the website
driver.get(url3)

time.sleep(1)

cross=driver.find_element_by_xpath('//button[@class="_2KpZ6l _2doB4z"]') #Closing the Pop-up that arrises everytime when we open flipkart website.
cross.click()

search = driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input') #given the xpath for the search button
search.send_keys(input())                                         #Used Input function 

click = driver.find_element_by_xpath('//button[@class="L0Z3Pu"]') #given the xpath for the click button
click.click()
time.sleep(2)
url = []                                                          #empty list
try:                                                              #Used try and except function
    ur = driver.find_elements_by_xpath('//div[@class="_2kHMtA"]/a')
    for i in ur:                                      
        url.append(i.get_attribute('href'))                      #Here I have extracted the url's for each of the product.
except NoSuchElementException:
        url.append("No Brand Available")
len(url)
```

```python
brand=[]                                #Empty list for the items what all I have to extract.
price=[]
product=[]
colour=[]
ram=[]
rom=[]
primary=[]
secondary=[]
display=[]
battery=[]

for i in url:

    try:                               #Used try and excpt so that just in case if we don't have the data under that particular class then also code should run without giving error. 
        driver.get(i)
        time.sleep(1)
        br=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        brand.append(br.text.split()[0])
        but=driver.find_element_by_xpath('//button[@class="_2KpZ6l _1FH0tX"]')
        but.click()
    except NoSuchElementException:
        brand.append("-")
        
    try:
        pri=driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
        price.append(pri.text)
    except NoSuchElementException:
        price.append("-")
        
    try:
        pro=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        product.append(pro.text)
    except NoSuchElementException:
        product.append("-")
        
    try:
        co=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        colour.append(co.text.split()[2:4])
    except NoSuchElementException:
        colour.append("-")
    
    try:
        ra=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[1]')
        ram.append(ra.text.split('|')[0])
    except NoSuchElementException:
        ram.append("-")
     
    try:
        ro=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[1]')
        rom.append(ro.text.split('|')[1])
    except NoSuchElementException:
        rom.append("-")
    
    try:
        pri=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[3]')
        primary.append(pri.text.split('|')[0])
    except NoSuchElementException:
        primary.append('-')
    
    try:
        sec=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[9]/div[4]/div/div[2]/div[1]/div[5]/table/tbody/tr[5]/td[2]/ul/li')
        secondary.append(sec.text)
    except NoSuchElementException:
        secondary.append('-')
    
    try:
        dis=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[2]')
        display.append(dis.text)
    except NoSuchElementException:
        display.append('-')
    
    try:
        bat=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[4]')
        battery.append(bat.text)
    except NoSuchElementException:
        battery.append('-')
        
print(len(brand))                     #Here I have printed the lenth of each of the data extracted so that we don't face any problem in creating a DataFrame.
print(len(price))
print(len(product))
print(len(ram))
print(len(rom))
print(len(colour))
print(len(primary))
print(len(secondary))
print(len(display))
print(len(battery))
```

# Ques 5
Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google
maps.

```python
url = 'https://www.google.co.in/maps' #opening the website
driver.get(url)

time.sleep(2)

search = driver.find_element_by_id("searchboxinput")                       # locating search bar
search.clear()                                                             # clearing search bar

time.sleep(2)

search.send_keys(input())                                                  # asking from the user to input the values in search bar

button = driver.find_element_by_id("searchbox-searchbutton")               # locating search button
button.click()                                                             # clicking search button

time.sleep(3)

try:

    url_string = driver.current_url                                       # we have extracted the current url in "url_string"

    print("URL Extracted: ", url_string)                                  # Url has to extracted because the latitude and longitude are in the url Itself.

    lat_lng = re.findall(r'@(.*)data',url_string)                         # Used RegEx Module (Regular Expressions) to extract latitude and longitude from the current url

    if len(lat_lng):

        lat_lng_list = lat_lng[0].split(",")                              # Spliting the Data and assiging the splitted data to "lat_lng_list"

        if len(lat_lng_list)>=2:                                          

            lat = lat_lng_list[0]                                          # Latitude

            lng = lat_lng_list[1]                                          # Longitude

        print("Latitude = {}, Longitude = {}".format(lat, lng))

except Exception as e:                                                  

        print("Error: ", str(e))
```

# Ques. 7
Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
url = 'https://www.digit.in/' #opening the website
driver.get(url)
time.sleep(2)
search = driver.find_element_by_xpath('/html/body/div[4]/div/div[2]/div[2]/div[4]/ul/li[9]/a')
search.click()
```

```python
heading = []
opera = []
display=[]
processor=[]
memory=[]

head=driver.find_elements_by_tag_name('h3') # Extracted the Entire Data of the laptops by the tag name("h3") 
for i in head:
    heading.append(i.text)                  # Name of laptops
os=driver.find_elements_by_xpath('//div[@class="product-detail"]/div/ul/li[1]/div')
for j in os:                                # Operating system used
    opera.append(j.text.split('\n')[1])
dis=driver.find_elements_by_xpath('//div[@class="product-detail"]/div/ul/li[2]/div')
for k in dis:                               # Display 
    display.append(k.text.split('\n')[1])
pro=driver.find_elements_by_xpath('//div[@class="product-detail"]/div/ul/li[4]/div/div')
for l in pro:                               # Processor  
    processor.append(l.text)
mem=driver.find_elements_by_xpath("//div[@class='Spcs-details'][1]/table/tbody/tr[6]/td[3]") 
for m in mem:                               # Memory
    memory.append(m.text)
                          
print(len(heading))                         # Printing the lenth
print(len(opera))
print(len(display))
print(len(processor))
print(len(memory))
```

```python
# Creating DataFrame
df6 = pd.DataFrame({'Name of Laptop':heading,'Operating System':opera,'Display':display,'Processor':processor,'Memory':memory})
df6
```

# Ques. 10
Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in 
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, 
overall reviews, privates from price, dorms from price, facilities and property description.

```python
url = 'https://www.hostelworld.com/'  # Opeing the url
driver.get(url)
time.sleep(2)
london=driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[7]/section/div/div[1]/a') # Clicking on london location
london.click()
time.sleep(5)
search = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[2]/section[1]/div/div/form/div/div[2]/button') # Clicking on search button to search all the hostel's in London location
search.click()
```

```python
name = []                        # Empty List
distance=[]
ratings=[]
reviews=[]
treview=[]
private=[]
dorms=[]
facilities=[]
hurl=[]

while(True):   
    nam=driver.find_elements_by_xpath('//h2[@class="title title-6"]')                # Extracting the Desired Element's by xpath
    dis=driver.find_elements_by_xpath('//div[@class="title-row"]/div/a/span[1]')
    rat=driver.find_elements_by_xpath('//div[@class="rating rating-summary-container big"]')
    rev=driver.find_elements_by_xpath('//div[@class="rating rating-summary-container big"]')
    tre=driver.find_elements_by_xpath('//div[@class="rating rating-summary-container big"]')
    pri=driver.find_elements_by_xpath('//div[@class="price-col"]')
    dor=driver.find_elements_by_xpath('//div[@class="price-col"]')
    faci=driver.find_elements_by_xpath('//div[@class="facilities-label facilities"]')
    urls=driver.find_elements_by_xpath("//h2[@class='title title-6']/a")

    for i in nam:
        name.append(i.text)                    # Appending the Text Data
    for j in dis:
        distance.append(j.text)
    for k in rat:
        ratings.append(k.text.split('\n')[0])
    for l in rev:
        reviews.append(l.text.split('\n')[1])
    for m in tre:
            treview.append(m.text.split()[2])
    for o in pri:
        private.append(o.text.replace('\n',''))
    for p in dor:
        dorms.append(p.text.replace('\n',''))
    for q in faci:
        facilities.append(q.text)
    for url in urls:
        hurl.append(url.get_attribute('href'))
        
    for n in reviews:                            # As there were some data in 'OverAll Reviews' column some of the data were missing, so replacing those data by "none"
        if "Rating"in n:
            reviews[reviews.index(n)] = "None"
        elif "Reviews"in n:
            reviews[reviews.index(n)] = "None"
       
    try:
        next_button = driver.find_element_by_xpath('//div[@class="pagination-item pagination-next"]')  #Clicking on next button

        next_button.click()

    except:

        break
print(len(name))                                # Printing the length of the  Extracted Data
print(len(distance))
print(len(ratings))
print(len(reviews))
print(len(treview))
print(len(private))
print(len(dorms))
print(len(facilities))
print(len(hurl))
```

Here "Privates from Price" and "Dorms from Price" were under the same class so there data were coming together, so just used range function and extracted the desired data that we want

```python
p=private[0:200:2]
pp=dorms[1:200:2]
print(len(p))
print(len(pp))
```

```python
# Extracting the Description of each hotels by individually going to each url of each hosel as we have already extracted the url of each Hostel's

description = []
for i in hurl:    
    try:
        driver.get(i)
        try:
            see=driver.find_element_by_xpath('//*[@id="__layout"]/div/div[1]/section/div[6]/div/div[2]/div/div/a')
            see.click()
        except:
            pass   
        time.sleep(1)
        des=driver.find_element_by_xpath('//div[@class="description-container"]')
        description.append(des.text)
    
    except NoSuchElementException:
        description.append("No Description Availabe")
```
