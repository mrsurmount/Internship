```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.imdb.com/list/ls056092300/')
page
```

```python
soup =  BeautifulSoup(page.content)
print(soup.prettify())
```

```python
movies = []
for i in soup.find_all('h3', class_="lister-item-header"):
    movies.append(i.text.replace('\n',""))   
movies
```

```python
ratings = []
for i in soup.find_all('div',class_="ipl-rating-star small"):
    ratings.append(i.text.replace('\n',""))   
ratings
```

```python
df = pd.DataFrame()
df['Top Rated 100 Indian movies and their Year of Release'] = movies
df['Ratings'] = ratings
df
```
