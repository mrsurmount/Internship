```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.nobroker.in/property/sale/bangalore/multiple?searchParam=W3sibGF0IjoxMy4wNjk4MjgxLCJsb24iOjc3LjU4OTYwNTQ5OTk5OTk5LCJwbGFjZUlkIjoiQ2hJSnowVW1UU0lZcmpzUm1CRFhNMnZXS3pzIiwicGxhY2VOYW1lIjoiSW5kcmEgTmFnYXIifSx7ImxhdCI6MTIuOTMwNzczNSwibG9uIjo3Ny41ODM4MzAyLCJwbGFjZUlkIjoiQ2hJSjJkZGxaNWdWcmpzUmgxQk9BYWYtb3JzIiwicGxhY2VOYW1lIjoiSmF5YW5hZ2FyIn0seyJsYXQiOjEyLjk5MjMyLCJsb24iOjc3LjUzMTAyMTUsInBsYWNlSWQiOiJDaElKZThPdGY3azlyanNSMDc4RVI3VG8wLUkiLCJwbGFjZU5hbWUiOiJSYWphcmFqZXNod2FyaSBOYWdhciJ9XQ==&radius=2.0&city=bangalore&locality=Indra%20Nagar,&locality=Jayanagar,&locality=Rajarajeshwari%20Nagar')
page
```

```python
soup = BeautifulSoup(page.content,"html.parser")
print(soup.prettify())
```

```python
heading = []
for i in soup.find_all('h2', class_= "heading-6 font-semi-bold nb__25Cl7",):
    heading.append(i.text)
heading
#len(heading)
```

```python
location = []
for i in soup.find_all('div', class_= "nb__1EwQz",):
    location.append(i.text)
location
#len(location)
```

```python
area = []
for i in soup.find_all('div', class_= "nb__FfHqA",):
    area.append(i.text)
#len(area)
area
```

```python
emi = []
for i in soup.find_all('div', class_= "font-semi-bold heading-6",id="roomType"):
    emi.append(i.text)
#len(emi)
emi
```

```python
price = []
for i in soup.find_all('div', class_= "nb__4L90a",itemprop="valueReference"):
    price.append(i.text)
#len(price)
price
```

```python
df = pd.DataFrame()
df['House Title'] = heading
df['Location'] = location
df['Area'] = area
df['EMI'] = emi
df['Price'] = price
df
```
