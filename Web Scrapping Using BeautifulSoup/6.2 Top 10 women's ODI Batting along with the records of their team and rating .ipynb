```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
page
```

```python
soup =  BeautifulSoup(page.content)
print(soup.prettify())
```

```python
player1 = []
for i in soup.find_all('div', class_="rankings-block__banner--name-large",):
    player1.append(i.text)
player1
```

```python
player2 = []
for i in soup.find_all('td', class_="table-body__cell rankings-table__name name",):
    player2.append(i.text.replace('\n',""))
players = player2[0:9]
players
```

```python
total_players = player1 + players
total_players
```

```python
team = []
for i in soup.find_all('div', class_="rankings-block__banner--nationality",):
    team.append(i.text.replace('\n',""))
team
```

```python
teams = []
for i in soup.find_all('span', class_="table-body__logo-text",):
    teams.append(i.text)
tt = teams[0:9]
tt
```

```python
total_team = team + tt
total_team
```

```python
rating = []
for i in soup.find_all('div', class_="rankings-block__banner--rating"):
    rating.append(i.text)
rating
```

```python
ratings = []
for i in soup.find_all('td', class_ = "table-body__cell rating"):
    ratings.append(i.text)
rati = ratings[0:9]
rati
```

```python
total_ratings = rating + rati
total_ratings
```

```python
df = pd.DataFrame()
df['Top_player'] = total_players
df['Team'] = total_team
df['Ratings'] = total_ratings
df
```
