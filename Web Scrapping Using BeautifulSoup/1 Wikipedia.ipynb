```python
from bs4 import BeautifulSoup
import requests
```

```python
wiki = requests.get('https://en.wikipedia.org/wiki/Main_Page')
wiki
```

```python
wikip = BeautifulSoup(wiki.content)
wikip
```

```python
header_tags = []
for header in wikip.find_all(["h1","h2","h3","h4","h5","h6"]):
    header_tags.append(header.text.strip())
header_tags
#len(header_tags)
```

```python
import pandas as pd
df = pd.DataFrame()
df['Heading']= header_tags
df
```
